```python
#Wärmekapazität
fluid = "MM"
T = 35 + 273.15   # K
P = 101325        # Pa

cp = PropsSI("C", "T", T, "P", P, fluid)
cp_mass = PropsSI("CPMASS", "T", T, "P", P, fluid)
cv = PropsSI("O", "T", T, "P", P, fluid)
cv_mass = PropsSI("CVMASS", "T", T, "P", P, fluid)
print(cp, cp_mass, cv, cv_mass, "J/kg/K")

cp_molar = PropsSI("CPMOLAR", "T", T, "P", P, fluid)
cv_molar = PropsSI("CVMOLAR", "T", T, "P", P, fluid)
print(cp_molar, cv_molar, "J/mol/K")
```

```python
fluids = ["Water", "MM", "MDM", "Isopentane"]

T = 35 + 273.15
P = 101325

for fluid in fluids:
    try:
        cp = PropsSI("CPMASS", "T", T, "P", P, fluid)
        cv = PropsSI("CVMASS", "T", T, "P", P, fluid)
        phase = PropsSI("PHASE", "T", T, "P", P, fluid)

        print(f"{fluid:10s} cp={cp:10.2f} J/kg/K   cv={cv:10.2f} J/kg/K   phase={phase}")

    except Exception as e:
        print(f"{fluid:10s} failed: {e}")
```
